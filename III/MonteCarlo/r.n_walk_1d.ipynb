```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(steps):
    position = 0
    positions = [position]
    for _ in range(steps):
        step = np.random.choice([-1, 1])  # Randomly choose direction: -1(left) or 1 (right)
        position += step
        positions.append(position)
    return positions

# Number of steps in the random walk
steps = 1000
# Perform random walk
positions = random_walk(steps)
# Plot the random walk
plt.plot(positions)
plt.title('1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def rw1D(n):
    x, y = 0, 0
    
    #Generate the time points [1,2,3,..., n]
    timepoints = np.arange(n+1)
    positions = [y]

    # Move the objects up or down
    for i in range(1, n + 1):

        #Randomly select either UP or DOWN
        step = random.random()

        # Move the object up or down
        if step <= 0.5:
            y += 1
        elif step > 0.5:
            y -= 1

        # Keep track of the positions
        positions.append(y) 
    return timepoints, positions

rw1 = rw1D(10000000)
rw2 = rw1D(10000000)
rw3 = rw1D(10000000)

plt.plot(rw1[0],rw1[1], "r-", label="rw1")
plt.plot(rw2[0],rw2[1], "g-", label="rw2")
plt.plot(rw3[0],rw3[1], "b-", label="rw3")
plt.show()
```
