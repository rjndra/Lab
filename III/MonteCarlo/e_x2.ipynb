```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f1(x):
    return np.exp(-x**2)

def f2(x):
    return np.exp(-x)

def f3(x):
    return f1(x)*f2(x)

def f4(x):
    return f1(x)/f2(x)

# Generate x values
x = np.linspace(-3, 3, 400)

# Generate y values
y1 = f1(x)
y2 = f3(x)
y3 = f4(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='$e^{-x^2}$', color='blue')
plt.plot(x, y2, label='$e^{-x^2}$', color='red')
plt.plot(x, y3, label='$e^{-x^2}$', color='yellow')
plt.title('Plot of $e^{-x^2}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
