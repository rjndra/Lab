```python
import numpy as np
import matplotlib.pyplot as plt
 
# Initialize parameters
N = 100  # Total number of molecules
steps = 100000  # Total number of steps
state = np.zeros(steps + 1)  # Number of particles in box 2 (initially empty)


urnA = list(range(1, N + 1))
urnB = []
count_urnA = []
count_urnB = []

mean_urnB = []
std_urnB = []
variance_urnB = []
error_urnb = []
 
# Simulation of the Markov chain
for step in range(1, steps + 1):
    # Select a molecule at random and move it
    if np.random.rand() < (state[step-1] / N):
        state[step] = state[step-1] - 1  # Move molecule from box 2 to box 1
    else:
        state[step] = state[step-1] + 1  # Move molecule from box 1 to box 2

    
    mean_urnB.append(np.mean(state))
    _var_A = np.var(urnA)
    var_urnA.append(_var_A)
 
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(state, label='Number of Molecules in Box 2')
plt.title('Ehrenfest Model: Molecules in Box 2 vs. Steps')
plt.xlabel('Step')
plt.ylabel('Number of Molecules in Box 2')
plt.legend()
plt.grid(True)
plt.show()

 
```
