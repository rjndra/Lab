```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(steps):
    x_position = 0
    y_position = 0
    x_cord = [x_position]
    y_cord = [x_position]
    for _ in range(steps):
        step = np.random.random()
        if step <= 0.25 : # (+1,0) -> if r.n < 0.25
            x_position += 1  
        elif 0.25 < step <= 0.5: # (-1,0) -> if r.n belongs (0.25,0.5)
            x_position -= 1 
        elif 0.5 < step <= 0.75: # (0,+1) -> if r.n belongs (0.5, 0.75)
            y_position += 1 
        elif step > 0.75: # (0,-1) -> if r.n > 0.75
            y_position -= 1  
        y_cord.append(y_position)
        x_cord.append(x_position)
    return x_cord,y_cord

# Number of steps in the random walk
steps = 100

# Perform random walk
x_cord, y_cord = random_walk(steps)

# Plot the random walk
plt.plot(x_cord, y_cord, color='blue')
plt.title('2D Random Walk')
plt.xlabel('x_position')
plt.ylabel('y_position')
plt.show()
```
