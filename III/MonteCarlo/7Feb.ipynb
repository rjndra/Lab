Integration: Algorithm

Choose the number of Monte Carlo samples N

Perform a lopp over N and for each generate a random number xi in the interval [0,1] through a
a call to a random number generator. Translate the random numbers to other required interval if it needs.

Use this number to evaluate f(xi)

Evaluate the conrtibutions to the mean value and the standard deviation fro each loop

After N samples calculate the final mean value and the standard deviation.

```python
import numpy as np
import math

seed = 23
N = 100000 # samples
rand = np.random.rand(N)
lowerlimit = 0
upperlimit = 1

dx = (upperlimit-lowerlimit)/N

def form(x):
    return pow(math.e,x)

fx = form(rand)

areas = (fx[:-1]+fx[1:]) * dx/2

integrals = sum(areas)
print(integrals)

import matplotlib.pyplot as plt
plt.subplot(2,1,1)
plt.plot(rand,fx)
plt.subplot(2,1,2)
plt.plot(range(N), fx)
plt.show()

def standardDeviation(fx):
    sd = 0
    N = len(fx)
    avg = sum(fx)/N
    for x in fx:
        val = x**2 - avg**2
        sd += val
    sdeviation = sd/N
    return sdeviation

sd = standardDeviation(fx)
print(f'Standard Deviation: {sd}')

se = []
for x in range(N):
    err = sd/math.sqrt(x)
    se.append(err)

print(f'Error {se}')
plt.plot(N,se)
```
