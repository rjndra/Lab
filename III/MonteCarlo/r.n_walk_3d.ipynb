```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(steps):
    x_position = 0
    y_position = 0
    z_position = 0
    x_cord = [x_position]
    y_cord = [x_position]
    z_cord = [z_position]
    for _ in range(steps):
        step = np.random.random()
        if step <= 0.15 : # (+1,0) -> if r.n < 0.25
            x_position += 1  
        elif 0.15 < step <= 0.3: # (-1,0) -> if r.n belongs (0.25,0.5)
            x_position -= 1 
        elif 0.3 < step <= 0.45: # (0,+1) -> if r.n belongs (0.5, 0.75)
            y_position += 1 
        elif 0.45 < step <= 0.6: # (-1,0) -> if r.n belongs (0.25,0.5)
            y_position -= 1 
        elif 0.6 < step <= 0.8: # (0,+1) -> if r.n belongs (0.5, 0.75)
            z_position += 1 
        elif step > 0.8: # (0,-1) -> if r.n > 0.75
            z_position -= 1  
        x_cord.append(x_position)
        y_cord.append(y_position)
        z_cord.append(z_position)
    return x_cord,y_cord, z_cord

# Number of steps in the random walk
steps = 1000

# Perform random walk
x_cord, y_cord, z_cord = random_walk(steps)

from mpl_toolkits import mplot3d
# Plot the random walk
axes = plt.axes(projection = '3d')

axes.plot3D(x_cord,y_cord,z_cord, 'green')
axes.set_title("3D Random Walk")
plt.show()
```
