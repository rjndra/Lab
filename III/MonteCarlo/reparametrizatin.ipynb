```python
numsamples = 1000000

x_samples_no_reparam = []
y_samples_no_reparam = []

x_samples_reparam = []
y_samples_reparam = []

# Without reparametrization

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(23)

def x_1(y_0):
    return np.random.uniform(y_0) # using standard normal distribution to get multivarite distribution.

def y_1(x_1):
    return np.random.uniform(x_1) # using standard normal distribution to get multivariate distribution.

u_no_resamp = np.random.uniform(-2,2, numsamples)
v_no_resamp = np.random.uniform(u_no_resamp, numsamples)

#Gibbs Sampling
for i in range(numsamples):
    x = x_1(v_no_resamp[i])
    y = y_1(x)
    x_samples_no_reparam.append(x)
    y_samples_no_reparam.append(y)
```

```python
plt.hist(y_samples_no_reparam, bins= 1000)
plt.show()
```

```python
# With Reparametrization

u = np.random.uniform(0,1, numsamples)
v = np.random.uniform(0,1, numsamples)

for i in range(numsamples):
    x = 4 * u[i] - 2
    y = 2 * v[i]*np.abs(x) - np.abs(x)
    x_samples_reparam.append(x)
    y_samples_reparam.append(y)
```

```python
plt.hist(y_samples_reparam, bins=1000)
plt.show()
```
