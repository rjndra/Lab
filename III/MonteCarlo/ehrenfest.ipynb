```python
import numpy as np
import matplotlib.pyplot as plt

def ehrenfest_model_simulation(num_particles, num_steps):
    container_a = num_particles // 2
    container_b = num_particles - container_a

    # Lists to store the number of particles in each container at each step
    container_a_history = [container_a]
    container_b_history = [container_b]

    for _ in range(num_steps):
        # Randomly select a particle and move it to the other container
        if np.random.rand() < 0.5:
            container_a -= 1
            container_b += 1
        else:
            container_a += 1
            container_b -= 1

        # Store the current state in the history lists
        container_a_history.append(container_a)
        container_b_history.append(container_b)

    return container_a_history, container_b_history

def plot_ehrenfest_simulation(container_a_history, container_b_history):
    steps = len(container_a_history)
    time = np.arange(steps)

    plt.plot(time, container_a_history, label='Container A')
    plt.plot(time, container_b_history, label='Container B')
    plt.xlabel('Time Steps')
    plt.ylabel('Number of Particles')
    plt.title('Ehrenfest Model Simulation')
    plt.legend()
    plt.show()

# Example usage
num_particles = 100
num_steps = 100000

container_a_history, container_b_history = ehrenfest_model_simulation(num_particles, num_steps)
plot_ehrenfest_simulation(container_a_history, container_b_history)
```
