```python
import random
import matplotlib.pyplot as plt
import numpy as np
random.seed = 15
N = 100
A_urn = list(range(1, N + 1))
B_urn = []
count_A = []
count_B = []
n = 100000
mean_A = []
var_A = []
sd_A = []
err_A = []
for i in range(1, n+1):
    ran = random.randint(1, N + 1)
    if ran in A_urn:
        # del A_urn[ran]
        A_urn.remove(ran)
        B_urn.append(ran)
    elif ran in B_urn:
        B_urn.remove(ran)
        A_urn.append(ran)
        
    mean_A.append(np.mean(A_urn))
    _var_A = np.var(A_urn)
    var_A.append(_var_A)
    
    _sd = _var_A ** 0.5
    sd_A.append(_sd)
    err_A.append(_sd / (i ** 0.5))
    
    count_A.append(len(A_urn))
    count_B.append(len(B_urn))
    
    
print("Final Balls in A :"+str(len(A_urn)))
print("Final Balls in B :"+str(len(B_urn)))
    
plt.plot(range(n), count_A, label="Balls in A")
plt.plot(range(n), count_B, label="Balls in B")
plt.title("Ehrenfest Model")
plt.xlabel("1000 steps")
plt.ylabel("Box count of A & B")
plt.legend()
plt.show()
mean_A_ = np.mean(A_urn)
var_A_ = np.var(A_urn)
sd_A_ = np.std(A_urn)
err_A_ = np.std(A_urn) / n ** 0.5
print("Mean is "+str(mean_A_))
print("Var is "+str(var_A_))
print("SD is "+str(sd_A_))
print("Error is "+str(err_A_))
plt.plot(range(n), err_A)
plt.show()
```
