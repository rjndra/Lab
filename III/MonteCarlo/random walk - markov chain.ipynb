```python
import random
import matplotlib.pyplot as plt
import numpy as np

random.seed(15)

x = 0
y = 0
z = 0

xl = []
yl = []

MSD = 0
MSDL = []
sqrt_i = []

n = 1000

for i in range(1, n):
    rn = random.random()
    
    if rn < 0.25:
        x += 1
        y += 0
    elif 0.25 < rn < 0.5:
        x -= 1
        y += 0
    elif 0.5 < rn < 0.75:
        x += 0
        y += 1
    else:
        x += 0
        y -= 1
    
    _MSD = x**2 + y**2
    MSD += _MSD
    
    MSDL.append(np.sqrt(MSD / (i) )) 
    
    sqrt_i.append(np.sqrt(i))
    
    xl.append(x)
    yl.append(y)

plt.show()

plt.plot(MSDL, range(1, n), color="green")
plt.plot(MSDL, sqrt_i, color="red")
plt.show()

MSD = np.sqrt(MSD / n)
print(MSD)

plt.plot(range(1, n), xl)
plt.show()

plt.plot(xl,yl)
plt.show()

ax = plt.axes(projection='3d')

# Data for a three-dimensional line
ax.plot3D(xl, yl, range(1, n), 'gray')
plt.show()
```
