```python
import nltk

word = "पात"
dictionary = ['पातलो', 'पात', 'पत्रु', 'पात्र', 'रामहरी']

for w in dictionary:
    ed = nltk.edit_distance(word, w)
    print(f'{word} <=> {w} = {ed}')
```

```python
from nltk.corpus import words
from nltk.util import ngrams

# from nltk.metrics.jaccard_distance import j_distance

correct_words = words.words()

incorrect_words = ['happyi', 'azaming', 'intellgent']

for word in incorrect_words:
    temp = [(nltk.edit_distance(word, w), w) 
            for w in correct_words if w[0] == word[0]] 
    sortedTemp = sorted(temp)
    print(sortedTemp)


print("Jaccard Distance")

for word in incorrect_words:
    ngramsOfWords =  set(ngrams(word, 2))
    temp = [(nltk.jaccard_distance(ngramsOfWords, set(ngrams(w,2))), w) 
            for w in correct_words if w[0] == word[0]] 
    sortedTemp = sorted(temp, key = lambda val:val[0])[0][1]
    print(sortedTemp)
```
