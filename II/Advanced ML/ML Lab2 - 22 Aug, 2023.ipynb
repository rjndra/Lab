```python
import numpy as np
np.random.seed(25)

#prepare toy dataset using equation y = 2x1 + 5 + e

x1 = np.random.randn(100)
y = 2 * x1 + 5 + np.random.randn(100) * 0.9
```

```python
#Plot the graph
import matplotlib.pyplot as plt
plt.scatter(x1, y)
plt.show()
```

```python
X = np.array(x1).reshape((-1,1))
print(X.shape)
```

```python
# Split train test dataset
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 25)

print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create Linear Regression Model

linear_model = LinearRegression().fit(train_X,train_y)
```

```python
# Predict on linear model
y_pred = linear_model.predict(test_X)

#plot the predicted outcome
plt.scatter(test_X, test_y, marker = ".")
plt.plot(test_X, y_pred, color = "orange")
plt.show()
```

```python
print("R2 : ", linear_model.score(test_X, test_y))
```

```python
print("Regression model coeffection are :" + str(linear_model.coef_[0]) + " and " + str(linear_model.intercept_))
```

```python
from sklearn.metrics import mean_squared_error

print("Mean squared error is " + str(mean_squared_error(test_y, y_pred)))
print("Root Mean Squared Error is " + str(np.sqrt(mean_squared_error(test_y, y_pred))))
```
