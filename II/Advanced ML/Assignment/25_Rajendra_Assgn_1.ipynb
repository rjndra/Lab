```python
import numpy as np
```

```python
#1 Create toy dataset
np.random.seed(25)

# Lets use eqation y = 3x1 + 9 
x1 = np.random.randn(100)
y = 3 * x1 + 9 + np.random.randn(100) * 0.8
```

```python
#2 Plot the give dataset
# for plot import matplotlib library
import matplotlib.pyplot as plt
plt.scatter(x1, y)
plt.show()
```

```python
# 3. Create a function for Linear Model

def Linear_Regression(w, b):
    y_hat = w * x1 + b
    return y_hat
```

```python
# 4 Create a function for Standard Scaler
def standard_Scaling(X):
    X = np.array(x1).reshape((-1, 1))
    return X
```

```python
# 5. Create a function for Computing loss - use MSE as loss function
def MSE(y_hat, y):
  return np.mean((y - y_hat)** 2)
```

```python
# 6. Train a model

X = standard_Scaling(x1)

# Frist the dataset
# Split given data into training data and test data
# Use sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=11)

# Initialize w & b
w = -1
b = -1

epochs = 100
alpha = 0.1

# Next tune the W & b based on loss
y_hat = Linear_Regression(w,  b)
plt.scatter(x1, y)
plt.plot(x1, y_hat)


for epoch in range(epochs):
  loss = MSE(y_hat,  y)
  dw = np.sum(np.dot((y_hat - y), x1))/2
  db = np.sum(y_hat - y)/2

  # Update W & b
  w -= alpha * dw/len(x1)
  b -= alpha * db/len(x1)

  y_hat = Linear_Regression(w, b)

  print("Loss in epoch {}: {}".format(epoch, MSE(y_hat, y)))
```
